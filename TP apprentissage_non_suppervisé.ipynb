```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=1000, centers=3,n_features=2,random_state=0)
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(X)
kmeans.predict(X)
```

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1],c=labels,s=50,cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c="black",s=200,alpha=0.5)
```

```python
inertia=[]
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,init='random',max_iter=500)  
  kmeans.fit(X)
  inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),inertia)
plt.xticks(range(1,11))
plt.xlabel('Number of cluster')
plt.ylabel('inertia')
plt.show()
```

4 est le nombre de cluster ideal

```python
model=KMeans(n_clusters=3,init='random',max_iter=500)  
model.fit(X)
```

```python
new_labels = model.labels_
new_centers = model.cluster_centers_
plt.scatter(X[:,0], X[:,1],c=new_labels,s=50,cmap='viridis')
plt.scatter(new_centers[:,0], new_centers[:,1], c="black",s=200,alpha=0.5)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(3)
agg.fit(X)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage

Z = linkage(X, 'ward')
dendrogram(Z,color_threshold=5)
plt.show()
```
